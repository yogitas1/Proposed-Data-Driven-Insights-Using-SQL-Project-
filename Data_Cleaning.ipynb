```python
file_columns = {}

file_names = ['game', 'game_info', 'game_summary', 'team', 'team_details', 'team_history', 'team_info_common', 'draft_history', 'draft_combine_stats']

for file_name in file_names:
  df = pd.read_csv(f'/content/{file_name}.csv')
  file_columns[file_name] = df.columns.tolist()

common_columns = {}
for file1 in file_columns:
  for file2 in file_columns:
    if file1 != file2:
      common = set(file_columns[file1]).intersection(file_columns[file2])
      if common:
        key = tuple(sorted((file1, file2)))
        if key not in common_columns:
          common_columns[key] = common

for (file1, file2), columns in common_columns.items():
  print(f"Common columns between {file1} and {file2}: {', '.join(columns)}")
```

```python

# Create a dictionary to store the DataFrames
dfs = {}
for file_name in file_names:
  dfs[file_name] = pd.read_csv(f'/content/{file_name}.csv')

# Iterate over the common_columns dictionary and merge DataFrames
for (file1, file2), columns in common_columns.items():
  if len(columns) > 0:
    merged_df = pd.merge(dfs[file1], dfs[file2], on=list(columns), how='inner')
    print(f"Merged {file1} and {file2} on columns: {', '.join(columns)}")
    print(merged_df.head())
```

```python
merged_files = []
for (file1, file2), columns in common_columns.items():
  if len(columns) > 0:
    merged_df = pd.merge(dfs[file1], dfs[file2], on=list(columns), how='inner')
    new_file_name = f"{file1}_{file2}_merged.csv"
    merged_df.to_csv(new_file_name, index=False)
    merged_files.append(new_file_name)

print("Merged files created:")
for file in merged_files:
  print(file)
```
